```python
from google.cloud import bigquery
import pydata_google_auth

credentials = pydata_google_auth.get_user_credentials(
    ['https://www.googleapis.com/auth/cloud-platform']
)

PROJECT_ID = "true-winter-488604-v2"

client = bigquery.Client(credentials=credentials, project=PROJECT_ID)

print("Authenticated Successfully using project:", PROJECT_ID)
```

```python
query_preview = """
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
"""

client.query(query_preview).to_dataframe()
```

The first five bike trips from the Austin Bikeshare dataset are displayed in the output. With information such as trip ID, start and end times, bike ID, duration, stations, and subscriber type, each row represents a single journey. This preview provides a brief synopsis of the dataset and the types of data that can be analyzed.

```python
query_avg = """
SELECT 
    start_station_name,
    AVG(duration_minutes) AS avg_trip_time
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY avg_trip_time DESC
"""

df_avg = client.query(query_avg).to_dataframe()
df_avg.head()
```

The results display the mean trip length (in minutes) for every starting station in the Austin Bikeshare dataset. The table is arranged in decreasing order, placing the stations with the longest average trips at the top. For instance, journeys beginning at "East 5th/Shady Ln" exhibit the longest average duration (approximately 197 minutes), whereas other locations such as "Lakeshore @ Austin Hostel" show shorter average trips (around 65 minutes). This assists in pinpointing which stations typically have longer or shorter averages for rides

```python
query_count = """
SELECT 
    start_station_name,
    COUNT(*) AS trip_count
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
ORDER BY trip_count DESC
"""

df_count = client.query(query_count).to_dataframe()
df_count.head()
```

The results present the total number of rides started at each station in the Austin Bikeshare dataset. The table is organized in decreasing order, featuring the most frequently utilized stations at the beginning. For example, "21st/Speedway @ PCL" documents the highest number of trips (108,559), indicating it is a popular departure point, while other stations like "26th/Nueces" have fewer trips (44,179). This helps identify the stations with the greatest number of trips
