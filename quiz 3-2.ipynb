```python
#Aashish ch# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Display plots inside notebook
%matplotlib inline
```

```python
# Load dataset from URL
url = "https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv"
iris = pd.read_csv(url)

# Display first 5 rows
iris.head()
```

```python
# Check shape (rows, columns)
iris.shape
# Get column names
iris.columns
# Summary statistics
iris.describe()
# Check data types and missing values
iris.info()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=iris, x="Class")
plt.title("Count of Each Iris Class")
plt.xlabel("Class")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(6,4))

plt.hist(iris["PedalLength"], bins=20, edgecolor='black')

plt.title("Histogram of Pedal Length")
plt.xlabel("Pedal Length")
plt.ylabel("Frequency")

plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(data=iris, x="Class", y="PedalLength")
plt.title("Box Plot of Pedal Length by Class")
plt.xlabel("Class")
plt.ylabel("Pedal Length")
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.violinplot(data=iris, x="Class", y="PedalLength")
plt.title("Violin Plot of Pedal Length by Class")
plt.xlabel("Class")
plt.ylabel("Pedal Length")
plt.show()
```
