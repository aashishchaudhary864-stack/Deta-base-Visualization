```python
import pandas as pa
import matplotlib.pyplot as plt
import seaborn as sns


df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Data_Sets/iris.csv')
```

```python
ax = sns.scatterplot(data = df, x = 'SepalLength', y = 'SepalWidth')
ax.set(title = "Length vs Width",
       xticks = [x for x in range(4,9,1)])
plt.show()
```

I failed to put the hue='Class' argument in the scatterplot, hence I was unable to obtain various colors in class. Seaborn used the default blue color for each point and treated all the data as a single group because no grouping variable was supplied.

```python
color_map = {
    'Iris-setosa': 'blue',
    'Iris-versicolor': 'green',
    'Iris-virginica': 'indigo'
}
for species, color in color_map.items():
    subset = df[df['Class'] == species]
    plt.scatter(subset['SepalLength'], subset['SepalWidth'], color=color, label=species)
plt.xlabel("Sepal Length")
plt.ylabel("Sepal Width")
plt.title("Length vs Width")
plt.legend()  
plt.show()
```

I plotted Sepal Length vs. Sepal Width using Matplotlib.

At first, all points were blue because I didnt assign colors to each flower type. Matplotlib uses one default color if you dont specify any.

To fix it, I created a color dictionary (blue, green, indigo), mapped each flower class to a color, and passed it to the c= parameter. This let each class appear in a different color.
